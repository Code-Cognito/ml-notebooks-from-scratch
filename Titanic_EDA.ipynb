# Data Cleaning
Lets check for missing and duplicates data and clean them.

```python
df.isnull().sum()
```

So We can see here.
* The age( Numeric feature has 177 can be filled with Median or mod )
* The cabin. It is a String or obj type. The reason for NaN is some people did not take the cabin.
* The Embarked is also an object type, and it has two Missing values.

Here we can just fill the missing values with median but there can be a chance that the lower class passengers, or males got more missing Age then other. This can effect our prediction. So to see the corelation of Pclass or Sex with the Age missingness, we should:
*   Create a feature named ageMissing. This will be an indicator of 0 and 1.
*   Then we will see for Pclass and Sex, How many times that indicator ticks.

```python
# creating ageMissing

df['ageMissing'] = df['Age'].isnull().astype(int)
```

```python
print("Now the  Columns of the data set are :", df.columns)
```

```python
df.head()
```

Now lets visualize the number of age missing by Pclass and Sex

```python
import seaborn as sns
```

```python
sns.countplot(x = df['Pclass'], hue= df['ageMissing'])
```

```python
sns.countplot(x = df['Sex'], hue= df['ageMissing'])
```

```python
sns.barplot(data=df, x='Pclass', y= 'ageMissing', hue= 'Sex')
```

So we can see that Missingness of Age is depending on the Pclass and Sex.

So we can get every groups median age, by the Pclass and the Sex of every passenger. Lets do it using group by function

```python
df.groupby(['Pclass', 'Sex'])['Age'].median()
```

So here we can see that here for every class we have different sex along with there meadians.

Now we can use transform function that will give use the meadian for every sell by group. We just have to pick the meadian where there the Age is missing.

If we check how the Age column looks like if all the passenger has meadian.

```python
df.groupby(['Pclass','Sex'])['Age'].transform('median').head()
```

Here we can see, though the data has different classes and Sex all over, this transform function can fill all the data by there Pclass and Sex.

----
So from here we can fill the data with fillna function very easily. But There is a slim chance that other features like SibSp and	Parch can also have effect on age's missing value. Lets find it out

```python
df.head()
```

First lets chech the distribution by groupby function for each SibSp then Parch

```python
df.groupby('SibSp')['ageMissing'].mean()
```

Here what we see that for different SibSp count we have different numbers of missingness parcentage. If we see for SibSp 4 and 5 we have no missing Age, But for 0 and 3 SibSp the missngness is 22% and 25%. But for 8 SibSp we have 100% missing. Means those Passengers who has 8 Siblings or Spouses, there age is totally missing.

---
Now lets see for Parch how the missingness is distributed

```python
df.groupby('Parch')['ageMissing'].mean()
```

Here we can see for Parch 0 to 2 we have missing Age. So lets use groupby using all of these features. Then fill the missing values with them. If any nan left then we will add them by groups of Pclass and Sex type.

```python
cleaned_df = df
```

```python
cleaned_df['Age'] = df['Age'].fillna(
    df.groupby(['Pclass', 'Sex', 'SibSp', 'Parch'])['Age'].transform('median')
)
```

Now lets check how many missing values are there.

```python
cleaned_df['Age'].isnull().sum()
```

there are 8 NaN is there. Lets fill them with Pclass and Sex group

```python
cleaned_df['Age'] = cleaned_df['Age'].fillna(
    cleaned_df.groupby(['Pclass', 'Sex'])['Age'].transform('median')
)
```

```python
cleaned_df['Age'].isnull().sum()
```

Now there is no null Age. And it is filled with different groups.
